```python
X_train = pd.read_pickle("/content/gdrive/My Drive/News-Classification-Project-main/ttrain_PCA_cols.pkl")
X_val = pd.read_pickle("/content/gdrive/My Drive/News-Classification-Project-main/tvalid_PCA_cols.pkl")
```

```python
print(X_train.shape)

X_train.head(2)
```

```python
## clean up indices and do one-hot encoding
dat_list = [X_train, X_val]

for news in dat_list:
    
    ## get rid of excess indices
    news.drop('Unnamed: 0', axis=1, inplace=True)
    news.drop('index', axis=1, inplace=True)
    news.drop('level_0', axis=1, inplace=True)
    
    ## One-hot encoding for category cols 
    news['politics'] = 0
    news.loc[news.category_col=='POLITICS', 'politics']=1

    news['wellness'] = 0
    news.loc[news.category_col=='WELLNESS', 'wellness']=1

    news['entertainment'] = 0
    news.loc[news.category_col=='ENTERTAINMENT', 'entertainment']=1

    news['parenting'] = 0
    news.loc[news.category_col=='PARENTING', 'parenting']=1

    news['style'] = 0
    news.loc[news.category_col=='STYLE & BEAUTY', 'style']=1

    news['travel'] = 0
    news.loc[news.category_col=='TRAVEL', 'travel']=1

    news['food'] = 0
    news.loc[news.category_col=='FOOD & DRINK', 'food']=1

    news['queer_voices'] = 0
    news.loc[news.category_col=='QUEER VOICES', 'queer_voices']=1

    news['business'] = 0
    news.loc[news.category_col=='BUSINESS', 'business']=1

    news['sports'] = 0
    news.loc[news.category_col=='SPORTS', 'sports']=1

    news['black_voices'] = 0
    news.loc[news.category_col=='BLACK VOICES', 'black_voices']=1

    news['world_news'] = 0
    news.loc[news.category_col=='WORLD NEWS', 'world_news']=1
```

```python
### Pick out columns which only have PCA Cols

pca_cols = X_train.columns[9:-12]
#X_train  = news_train

pca_cols_v = X_val.columns[9:-12]
#X_val = news_val

X_train[pca_cols]
```

```python
### y coloums: category names
cat_cols = X_train.columns[-12:]
y_train = X_train[cat_cols]

cat_cols_v = X_val.columns[-12:] 
y_val = X_val[cat_cols_v ]

y_train.head(2)
```

```python
## Train_test_split 
from sklearn.model_selection import train_test_split

X_ttrain, X_ho, y_ttrain, y_ho = train_test_split(X_train[pca_cols],y_train, 
                                         random_state = 546,
                                         shuffle = True,
                                         test_size = .1,
                                         stratify = X_train['category_col'])
```

## Model: Boosted Decision Trees 

### Tuning the hyperparameters - round 1

```python
max_depths = [10, 50, 100]
child_weights = [0.5, 1, 2]

cl_accs = np.zeros((len(max_depths), len(child_weights)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for weight in child_weights:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         sampling_method="gradient_based",
                                         learning_rate=.05,
                                         max_depth=max_depth,
                                         min_child_weight=weight,
                                         n_estimators = 200)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 10,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, weight]

          k = k + 1
    j = j + 1 
```

```python
## tuning stopped early, because max_depth of >30 not allowed.
## Here is accuracy score for models of with max_depth=10, 
## respectively, child_weights = 0.5, 1, 2
cl_accs
```

```python
max_depths = [20, 25, 30]
child_weights = [0.5, 1, 2]

cl_accs = np.zeros((len(max_depths), len(child_weights)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for weight in child_weights:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         sampling_method="gradient_based",
                                         learning_rate=.05,
                                         max_depth=max_depth,
                                         min_child_weight=weight,
                                         n_estimators = 200)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 10,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, weight]

          k = k + 1
    j = j + 1 
```

### Tuning the hyperparameters - round 3

```python
max_depths = [8, 10, 15]
child_weights = [1.5, 2, 3]

cl_accs = np.zeros((len(max_depths), len(child_weights)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for weight in child_weights:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         sampling_method="gradient_based",
                                         learning_rate=.05,
                                         max_depth=max_depth,
                                         min_child_weight=weight,
                                         n_estimators = 250)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 10,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)
          
          print(f"Max_depth: {max_depths[j]} and child_weight {child_weights[k]}: accuracy is {cl_accs[j,k] * 100}%")

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, weight]

          k = k + 1
    j = j + 1 
```

```python
# Get best parameters, plus the accuracy score for that combination
print(f"""Best accuracy is {best_acc * 100}%
Best parameters are {best_params}""")
```

columns represent min_child_weight (1.5, 2, 3)

### Tuning the hyperparameters, round 4

min_child_weight has held realtively stable at 2 (but also, the difference it makes does not increase the accuracy by a full percentage point), so for the next round, I'm going to focus on the max_depth parameter, and see if a bigger sample size will improve accuracy.

Note that in previous rounds of tuning, the lower max_depth was, the more estimators were needed, so I'm increasing n_estimators

```python
max_depths = [6, 8]
sample_size = [0.2, 0.5]

cl_accs = np.zeros((len(max_depths), len(sample_size)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for sample in sample_size:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=sample,
                                         sampling_method="gradient_based",
                                         learning_rate=.05,
                                         max_depth=max_depth,
                                         min_child_weight=2,
                                         n_estimators = 500)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 10,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)
          
          print(f"Max_depth: {max_depths[j]} and subsample {sample_size[k]}: accuracy is {cl_accs[j,k] * 100}%")

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, sample]

          k = k + 1
    j = j + 1 
```

```python
# Get best parameters, plus the accuracy score for that combination
print(f"""Best accuracy is {best_acc * 100}%
Best parameters are {best_params}""")
```

```python
cl_accs*100
```

Changing max_depth to 6 had the accuracy jump to ~57.43%. now going to just tune max depth

### Training the hyperparameters, round 5

```python
max_depths = [4,5,7]

cl_accs = np.zeros(len(max_depths))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    print(f"max_depth: {max_depths[j]}")
    xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                    subsample=.2,
                                    sampling_method="gradient_based",
                                    learning_rate=.05,
                                    max_depth=max_depth,
                                    min_child_weight=2,
                                    n_estimators = 600)

    xgb_cl.fit(X_ttrain, y_ttrain,
        early_stopping_rounds = 20,  
        eval_set=[(X_ho, y_ho)])

    pred = xgb_cl.predict(X_ho)

    cl_accs[j] = accuracy_score(y_ho,  pred)
    
    print(f"Max_depth: {max_depths[j]}: accuracy is {cl_accs[j] * 100}%")

    if cl_accs[j] > best_acc:
        best_acc = cl_accs[j]
        best_params = [max_depth]

    j = j + 1 
```

```python
# Get best parameters, plus the accuracy score for that combination
print(f"""Best accuracy is {best_acc * 100}%
Best parameters are {best_params}""")
```

### Training the hyperparameters, round 7

```python
max_depths = [4, 5]
learn_rate = [0.1, 0.3]

cl_accs = np.zeros((len(max_depths), len(learn_rate)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for rate in learn_rate:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         sampling_method="gradient_based",
                                         learning_rate=rate,
                                         max_depth=max_depth,
                                         min_child_weight=2,
                                         n_estimators = 1000)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 20,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)
          
          print(f"Max_depth: {max_depths[j]} and learning_rate {learn_rate[k]}: accuracy is {cl_accs[j,k] * 100}%")

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, rate]

          k = k + 1
    j = j + 1 
```

```python
print(f"""With a learning rate of 0.1, the accuracy of models with max_depth 4 and 5 (respectively) is:
 {np.around(cl_accs[0,0]*100, 2)} and {np.around(cl_accs[1, 0]*100, 2)}

 With a learning rate of 0.3, the accuracy of models with max_depth 4 and 5 (respectively) is:
 {np.around(cl_accs[0,1]*100, 2)} and {np.around(cl_accs[1,1]*100, 2)} 
 
 """)

# Get best parameters, plus the accuracy score for that combination
print(f"""Best accuracy is {np.around(best_acc * 100,2)}%
Best parameters are {best_params}""")
```

best accuracy barely worse than max_depth 4 with learning_rate 0.05, AND it got there with almost half the estimators (and had stabalized by that point!)

### Training the hyperparameters, round 8

```python
max_depths = [3, 4]
learn_rate = [.06, .07, .08]

cl_accs = np.zeros((len(max_depths), len(learn_rate)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for rate in learn_rate:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         sampling_method="gradient_based",
                                         learning_rate=rate,
                                         max_depth=max_depth,
                                         min_child_weight=2,
                                         n_estimators = 1000)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 20,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)
          
          print(f"Max_depth: {max_depths[j]} and learning_rate {learn_rate[k]}: accuracy is {np.around(cl_accs[j,k] * 100, 2)}%")

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, rate]

          k = k + 1
    j = j + 1 
```

### final final

looking at gamma (was a flop - also tried it with higher gammas)

```python
max_depths = [3]
gammas = [1, 5]

cl_accs = np.zeros((len(max_depths), len(gammas)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for g in gammas:
          print(j,k)
          xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                                         subsample=0.2,
                                         gamma=g,
                                         sampling_method="gradient_based",
                                         learning_rate=0.9,
                                         max_depth=max_depth,
                                         min_child_weight=2,
                                         n_estimators = 1000)

          xgb_cl.fit(X_ttrain, y_ttrain,
              early_stopping_rounds = 20,  
              eval_set=[(X_ho, y_ho)])

          pred = xgb_cl.predict(X_ho)

          cl_accs[j,k] = accuracy_score(y_ho,  pred)
          
          print(f"Max_depth: {max_depths[j]} and gamma {gammas[k]}: accuracy is {np.around(cl_accs[j,k] * 100, 2)}%")

          if cl_accs[j,k] > best_acc:
              best_acc = cl_accs[j,k]
              best_params = [max_depth, g]

          k = k + 1
    j = j + 1 
```

# Making the model

```python
#Make the model with these parameters
xgb_cl = xgboost.XGBClassifier(tree_method="gpu_hist",
                               subsample=0.2,
                               sampling_method="gradient_based",
                               learning_rate=.09,
                               max_depth=3,
                               min_child_weight=2,
                               n_estimators = 1000)

xgb_cl.fit(X_train[pca_cols], y_train,
   early_stopping_rounds = 20,  
   eval_set=[(X_train[pca_cols], y_train), (X_val[pca_cols_v], y_val)])
```

```python
# make predictions for val data
y_pred = xgb_cl.predict(X_val[pca_cols_v])
predictions = [value for value in y_pred]

# evaluate predictions 
accuracy = accuracy_score(y_val, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

f1 = f1_score(y_val, predictions, average='weighted')
print("f1 score: %.2f%%" % (f1 * 100.0))
```

### Plot the logloss scores

```python
# Plot the logloss 

results = xgb_cl.evals_result()

plt.figure(figsize=(10,7))
plt.plot(results["validation_0"]["logloss"], label="Training loss")
plt.plot(results["validation_1"]["logloss"], label="Validation loss") 
plt.xlabel("Number of trees", fontsize = 16)
plt.ylabel("Loss", fontsize = 16)
plt.legend()
plt.title("Model 1", fontsize = 20)

plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, predictions))
```
