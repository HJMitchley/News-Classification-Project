```python
def get_wordnet_pos(treebank_tag):
    ##### This function converts part of speech tags to a form the lemmatizer can use
    if treebank_tag.startswith('J'):
        return 'a'
    elif treebank_tag.startswith('V'):
        return 'v'
    elif treebank_tag.startswith('N'):
        return 'n'
    elif treebank_tag.startswith('R'):
        return 'r'
    else:
        return 'n'
```

```python
def preprocess_text(tstring):
    temp=''
    
    ## tokenize the words from the string
    tlist=nltk.word_tokenize(tstring.replace('-', ' '))
    
    ## remove punctation
    tlist =[w.lower() for w in tlist if (w not in string.punctuation) and (w not in ['\u2019', '\'s'])]
    
    ## convert numbers to words
    #tlist=[num2words(w) if w.isnumeric() else w for w in tlist]
    
    ## Tag part of speech
    tlist=nltk.pos_tag(tlist)
    
    ## convert to lowercase and remove stop words
    tlist=[w for w in tlist if not w[0] in set(stopwords.words('english'))]

    for word in tlist:
        #print(word[0], get_wordnet_pos(word[1]))
        if len(temp)!=0:
            #print(WordNetLemmatizer().lemmatize(word[0].lower(),get_wordnet_pos(word[1])))
            temp += ' ' + WordNetLemmatizer().lemmatize(word[0].lower(),get_wordnet_pos(word[1]))
        else: 
            #print(WordNetLemmatizer().lemmatize(word[0].lower(),get_wordnet_pos(word[1])))
            temp += WordNetLemmatizer().lemmatize(word[0].lower(),get_wordnet_pos(word[1]))
    return temp
```

```python
## Clean headlines
df['short_description_cleaned']=df['short_description_cleaned'].apply(lambda x: preprocess_text(x))
df['headline_cleaned']=df['headline_cleaned'].apply(lambda x: preprocess_text(x))
```

```python
## create new column with two cleaned cells concatenated
df['cleaned_words'] = df['headline_cleaned'] + ' ' + df['short_description_cleaned']
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
## Calcuate TF-IDF and create vectors
v = TfidfVectorizer()
x = v.fit_transform(df['cleaned_words'])
```

```python
## Save the list of vectors as a df, along with the words each point corresponds to
df_vectors = pd.DataFrame(x.toarray(), columns=v.get_feature_names_out())

## rename columns in original df 
## (it causes problems if one of the words from text is same as another column name)
df = df.rename(columns={"link":"link_col", 
                        "headline":"headline_col", 
                        "category":"category_col",
                        "authors":"authors_col", 
                        "date":"date_col"})

## concatenate the two dfs on the vertical axis
whole_df = pd.concat([df, df_vectors], axis = 1)
```
