```python
cat_cols = news.columns[-11:]
cat_cols
```

```python
### Columns which only have PCA Cols
pca_cols  = news.columns[9:-11]
pca_cols
```

```python
small_news = news.head(500)
```

```python
## Train_test_split
from sklearn.model_selection import train_test_split

X_train, X_ho, y_train, y_ho = train_test_split(small_news[pca_cols],small_news[cat_cols],
                                         random_state = 546,
                                         shuffle = True,
                                         test_size = .2,
                                         stratify = small_news['category_col'])
```

```python
X_eval, y_eval = train_test_split(small_news[pca_cols],small_news[cat_cols],
                                         random_state = 546,
                                         shuffle = True,
                                         test_size = .2,
                                         stratify = small_news['category_col'])
```

```python
## put in stuff for gridsearch
```

```python
### Base model, random parameters
xgb_cl = xgboost.XGBClassifier(learning_rate=.1,
                                   max_depth=8,
                                   n_estimators=40)


## fit it
xgb_cl.fit(X_train, y_train,
           early_stopping_rounds = 10,  
           eval_set=[(X_train, y_train), (X_ho, y_ho)])
```

## Fine tuning with for loops

```python
max_depths = [7, 9, 11]
child_weights = [0.5, 1, 2]
max_leaves = [0, 20, 50, 100]

cl_accs = np.zeros((len(max_depths), len(child_weights), len(max_leaves)))

best_acc = 0
best_params =  None

j = 0
for max_depth in max_depths:
    k = 0
    for weight in child_weights:
        i = 0
        for leaf in max_leaves:
            print(j,k,i)
            xgb_cl = xgboost.XGBClassifier(learning_rate=.05,
                                           max_depth=max_depth,
                                           min_child_weight=weight,
                                           max_leaves = leaf,
                                           n_estimators = 200)

            xgb_cl.fit(X_train, y_train,
               early_stopping_rounds = 10,  
               eval_set=[(X_ho, y_ho)])

            pred = xgb_cl.predict(X_ho)

            cl_accs[j,k,i] = accuracy_score(y_ho,  pred)

            if cl_accs[j,k,i] > best_acc:
                best_acc = cl_accs[j,k,i]
                best_params = [max_depth, weight, leaf]
            
            i = i + 1
        k = k + 1
    j = j + 1 
```

```python
print(f"""Best accuracy is {best_acc}
Best parameters are {best_params}""")
```

Ditch gridsearch, use forloops

Max_depth (7,9, 11) And min_child_weight (0.5, 1, 2) 

only pick 3 or 4 values for each, and train on those. If optimal value is on edge of range, recenter it on that optimal value
