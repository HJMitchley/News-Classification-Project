```python
%%time
##### This cell should fit the pca in batches #####
from gensim.matutils import corpus2dense
from sklearn.decomposition import IncrementalPCA
pca = IncrementalPCA(n_components=5000) ###n_components will determine the number of dimensions to keep after applying dimensionality reduction.
batch_size=5100 ###If you set the batch size to 1, it will throw an error, since the slice will no longer be a list of lists
n=int(len(train_corpus)/batch_size)-1
for i in range(n):
    print('Percent completion: ' + "{:10.2f}".format(100*i*batch_size/len(train_corpus)) + '%')
    #####Temp is a numpy array corresponding to the dense rows generated from the gensim tfidf
    temp = corpus2dense(train_tfidf[train_corpus[batch_size*i:batch_size*(i+1)]], num_terms=len(train_dict)).transpose()
    #####We incrementally fit the pca in batches
    pca.partial_fit(temp)

#####last batch has irregular size
temp = corpus2dense(train_tfidf[train_corpus[batch_size*n:]], num_terms=len(train_dict)).transpose()
pca.partial_fit(temp)
    
```

```python
pca.explained_variance_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
import math
#####This cell should generate the transformed data in batches
i=0
#####
for i in range(math.ceil(len(train_corpus)/batch_size)):
    print('Percent completion: ' + "{:10.2f}".format(100*i*batch_size/len(train_corpus)) + '%')
    temp = corpus2dense(train_tfidf[train_corpus[min(batch_size*i, len(train_corpus)):min(batch_size*(i+1), len(train_corpus))]], num_terms=len(train_dict)).transpose()
    print(temp.shape)
        
    #####out is a matrix containing the dimension reduced columns. Output to a pandas df not yet implemented. 
    out=pca.transform(temp)
    if i==0:
        outdf=pd.DataFrame(out)
    else:
        outdf=pd.concat([outdf, pd.DataFrame(out)], axis=0)
    
```

```python
#sanity check
print(outdf.shape)
print(len(train_corpus))
print(news_ttrain.shape)
```

```python
#####combine PCA output with original dataframe and save to a CSV
news_train=news_train.reset_index()
outdf=outdf.reset_index()
to_save=pd.concat([news_train, outdf], axis=1)
print('concatenation is done!')
to_save.to_pickle('/content/gdrive/My Drive/train_PCA_cols.pkl')
```
